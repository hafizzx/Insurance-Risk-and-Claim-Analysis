# **KPI’s Requirements:**

---



<h1>Total Policies

```python
total_policies = df['ID'].count()
print('Total Policies are:',total_policies)
```

<h1>Total Claim Amount

```python
total_clain_amount = df['Claim Amount'].sum()
print('Total Claim Amount is:',total_clain_amount)
```

<h1>Claim Frequency

```python
claim_frequency = df['Claim Freq'].sum()
print('Claim Frequency is:',claim_frequency)
```

<h1>Average Claim Amount

```python
avg_claim_amount = df['Claim Amount'].mean()
print('Average Claim Amount is:',avg_claim_amount)
```

<h1>Gender-wise Total Policies

```python
gender_policies = df.groupby('Gender')['ID'].count()
print('Gender-wise Total Policies are:\n',gender_policies)
```

```python
gender_policies.plot(kind="bar", color=["skyblue", "lightgreen"])
plt.title("Gender-wise Policy Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Policies")
plt.show()
```

<h1>By Car Use (Donut Chart)

```python
car_use_summary = df.groupby("Car Use").agg(
    Policies=('ID', 'nunique'),
    Claim_Amount=('Claim Amount', 'sum')
).reset_index()
car_use_summary
```

```python
car_use_summary.set_index("Car Use")["Policies"].plot.pie(
    autopct='%1.1f%%', startangle=90, figsize=(6,6), ylabel=""
)
plt.title("Policies by Car Use")
plt.show()
```

<h1>By Car Make (Bar Chart) -Bar chart can not work because there is lot of Makers

```python
car_make_summary = df.groupby("Car Make").agg(
    Policies=('ID', 'nunique'),
    Claim_Amount=('Claim Amount', 'sum')
).reset_index()
car_make_summary
```

```python
car_make_summary.plot(x="Car Make", y="Policies", kind="bar", figsize=(12,6))
plt.title("Policies by Car Make")
plt.xticks(rotation=45, ha="right")   # Rotate labels for readability
plt.tight_layout()
plt.show()
```

<h1>By Coverage Zone (Donut Chart)

<h1>By Age Group (Histogram)

```python
df['Age'] = pd.to_datetime("today").year - df['BirthDate'].dt.year

# Age Groups
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ["18-25", "26-35", "36-45", "46-55", "56-65", "65+"]
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_summary = df.groupby("Age Group").agg(
    Policies=('ID', 'nunique'),
    Claim_Amount=('Claim Amount', 'sum')
).reset_index()
age_summary
```

```python
age_summary.plot(x="Age Group", y="Policies", kind="bar", figsize=(8,5))
plt.title("Policies by Age Group")
plt.show()
```

<h1>Education & Marital Status (Matrix Heat Grid)

```python
matrix_summary = df.groupby(["Education", "Marital Status"]).agg(
    Policies=('ID', 'nunique'),
    Claim_Amount=('Claim Amount', 'sum')
).reset_index()
matrix_summary
```

```python
pivot_matrix = matrix_summary.pivot(
    index="Education", columns="Marital Status", values="Policies"
).fillna(0)

plt.figure(figsize=(8,6))
sns.heatmap(pivot_matrix, annot=True, fmt="g", cmap="Blues")
plt.title("Policies by Education & Marital Status")
plt.show()
```
